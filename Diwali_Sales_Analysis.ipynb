```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
plt.style.use('dark_background')
```

```python
df = pd.read_csv("Diwali Sales Data.csv",encoding='unicode_escape') # For avoiding encoding error
df.head()
```

```python
df.shape
```

```python
df.info()
```

## Dropping Status and unnamed1 column

```python
df.drop(["Status","unnamed1"],axis=1,inplace=True)
```

```python
df.isnull().sum() # Checking NULL values
```

```python
df.dropna(inplace=True)
```

```python
df.shape
```

```python
df.isnull().sum()
```

```python
df.info()
```

```python
df["Amount"] = df["Amount"].astype("int")
```

```python
df["Amount"].dtypes
```

```python
df.columns
```

```python
df.rename(columns={"Cust_name":"Customer_name"},inplace=True)
```

```python
df.head()
```

```python
df.describe()
```

```python
df[["Age","Orders","Amount"]].describe()
```

## Exploratory Data Analysis

### Gender

```python
ax = sns.countplot(x="Gender",data=df,palette="bright")

for bars in ax.containers:
    ax.bar_label(bars)  
```

```python
df.groupby(["Gender"],as_index=False)["Amount"].sum().sort_values(by="Amount",ascending=False)
```

```python
genderwise_sales = df.groupby(["Gender"],as_index=False)["Amount"].sum().sort_values(by="Amount",ascending=False)
```

```python
salesgraph = sns.barplot(x="Gender",y="Amount",data = genderwise_sales,palette="bright")

for i,j in enumerate(genderwise_sales["Amount"]):
    salesgraph.text(i,j,str(j),ha="center")
```

## Most Buyers are Female and purchasing power of female are greater than men

## Age

```python
df.columns
```

```python
ax = sns.countplot(data=df,x="Age Group",hue = "Gender",palette="bright")

for bars in ax.containers:
    ax.bar_label(bars)
```

## Female belonging to 26-35 Age group shop the most

```python
sales_byage = df.groupby(["Age Group"],as_index=False)["Amount"].sum().sort_values(by="Amount",ascending=False)
```

```python
sales_byage
```

```python
ax = sns.barplot(data=sales_byage,x="Age Group",y="Amount",palette="bright")

for i,j in enumerate(sales_byage["Amount"]):
    ax.text(i,j,str(j),ha="center")
```

## State

```python
# Orders from top 10 states
top10state = df.groupby(["State"],as_index = False)["Orders"].sum().sort_values(by="Orders",ascending  = False).head(10)
#top10state
```

```python
sns.set(rc={"figure.figsize":(15,5)})

ax = sns.barplot(data=top10state,x="State",y="Orders",palette="bright")

for i,j in enumerate(top10state["Orders"]):
    ax.text(i,j,str(j),ha="center")
```

```python
# Amount for top 10 states
amttop10state = df.groupby(["State"],as_index = False)["Amount"].sum().sort_values(by="Amount",ascending  = False).head(10)
```

```python
sns.set(rc={"figure.figsize":(15,5)})

ax = sns.barplot(data=amttop10state,x="State",y="Amount",palette="bright")

for i,j in enumerate(amttop10state["Amount"]):
    ax.text(i,j,str(j),ha="center")
```

## Most orders are from Uttar Pradesh , Maharashtra , Karnataka

## Martial Status

```python
mstatus = df.groupby(['Marital_Status','Gender'],as_index=False)["Amount"].sum().sort_values(by="Amount",ascending=False)
```

```python
sns.set(rc={"figure.figsize":(5,5)})

ax = sns.barplot(data=mstatus,x="Marital_Status",y="Amount",palette="bright",hue="Gender")
ax
```

## Most Shopping done by unmarried women

## Occupation

```python
occu = df.groupby(["Occupation"],as_index = False)["Amount"].sum().sort_values(by="Amount",ascending  = False)
```

```python
sns.set(rc={"figure.figsize":(20,5)})
sns.barplot(data=occu,x="Occupation",y="Amount",palette="bright")
```

## Most buyers are from IT Sectors , Healthcare , Aviation

## Product Category

```python
df.columns
```

```python
orders_bycategory = df.groupby(["Product_Category"],as_index = False)["Orders"].sum().sort_values(by="Orders",ascending=False) 
```

```python
ax = sns.barplot(data=orders_bycategory,x="Product_Category",y="Orders",palette="bright")

for i,j in enumerate(orders_bycategory["Orders"]):
    ax.text(i,j,str(j),ha="center")
```

```python
cate_byamt = df.groupby(["Product_Category"],as_index = False)["Amount"].sum().sort_values(by="Amount",ascending=False).head(10) # Top 10 Selling Categories 
```

```python
# Top 10 Selling Categories 
ax = sns.barplot(data=cate_byamt,x="Product_Category",y="Amount",palette="bright")

for i,j in enumerate(cate_byamt["Amount"]):
    ax.text(i,j,str(j),ha="center")
```

## Most of the products are from Food , Clothing and Apparel , Electronics and Gadgets category

## Sales by Product ID

```python
orders_byProdID = df.groupby(["Product_ID"],as_index=False)["Orders"].sum().sort_values(by="Orders",ascending=False).head(10)
```
